```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
# import random
# random.seed(0)
```

```python
graph = nx.erdos_renyi_graph(
    n = 20, 
    p = 0.3, 
    directed = False
)
```

```python
plt.figure(dpi = 300);
plt.figure(figsize=(10, 8));
plt.axis('off');
nx.draw_networkx(
    G = graph,
    # pos = nx.spectral_layout(graph), # Position nodes using the eigenvectors of the graph Laplacian.
    pos = nx.spring_layout(graph), 
    node_size = 600,
    cmap = 'coolwarm',
    font_color = 'white',
    font_size = 16
);
```

```python
start, end = 1, 2
```

```python
print(f"Shortes Path between {start} and {end} is {nx.shortest_path(graph, start, end)}")
```
